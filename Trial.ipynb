```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 500, 0.1)
```

```python
v_max = 40 # assumed
eta_p = 0.8 # assumed sadraey
AR = 9 # similiar aircrafts have this ()
e = 0.8 #sadraey
rho_0 = 1.225 # standard atm
cd_0 = 0.024 #sadraey
k = 1/ (np.pi*e*AR)
sigma = 1 #electric so power ratio is 1
a = 0.5 * rho_0 * cd_0
b = 2 * k/(rho_0*sigma)
rho_1000 = 1.112
L_D_max = 15 #
roc = 2.5
```

```python
def plot(v_max, eta_p, AR, e, rho_0, cd_0, k, sigma, a, b, rho_1000, L_D_max, roc):
  y1 = eta_p/(a*v_max**3/x + b*x/v_max)
  y2 = 1/(100/(3*eta_p) + np.sqrt(2*x/(rho_1000*np.sqrt(3*cd_0/k)))*1.155/(L_D_max*eta_p))
  y3 = 1/ (roc/eta_p + np.sqrt(2 / (rho_0* np.sqrt(3*cd_0/k)) * x )*(1.155 / (L_D_max*eta_p)) )
  plt.plot(x, y1, label = "max_speed")
  plt.plot(x, y2, label= "celing")
  plt.plot(x, y3, label= "ROC")
  plt.grid(True)
  plt.xlabel("W/S")
  plt.ylabel("W/P")
  plt.legend()
```

```python
plot(v_max, eta_p, AR, e, rho_0, cd_0, k, sigma, a, b, rho_1000, L_D_max, roc)
```
